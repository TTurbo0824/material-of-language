```python
from scipy.stats import truncnorm
def t_normal(a, b, mu, sigma):
    tn = truncnorm((a-mu)/sigma, (b-mu)/sigma, loc=mu, scale=sigma)
    return tn.rvs(1)[0]
```

## Interactive Widget

Example

```python
def disp_square(size=40):
    page = document(80, 80, 'mm').addpage()
    pen = shape().fill(rgb(120, 40, 160)).nostroke()
    page.place(pen.rectangle(10, 10, size, size))
    show(page)
```

```python
size_slider = widgets.IntSlider(min=0, max=70, value=35)
widgets.interact(disp_square, size=size_slider)
```

#### Polyline function

'a' should be changing according to the number of strokes of each letter and 'b' should be depending on the slider bar

```python
# Add 0 for any punctuation marks or space
alpha_s = [3, 3, 1, 2, 4, 4, 2, 3, 3, 2, 3, 2, 4, 3, 1, 2, 2, 3, 1, 2, 1, 2, 4, 2, 3, 3]

# Checking the length of array
len(alpha_s)
```

```python
# Not interactive
def make_char(a):
    pts = []
    for i in range(int(t_normal(2*a+3, 2*a+4, 8, 2))):
        x = (choice([0, 0.5, 1]))
        y = (choice([0, 0.5, 1]))
        pts.append([x, y])
    return Polyline(pts)
```

```python
# Interactive
def make_char(a, b = 3):
    pts = []
    for i in range(int(t_normal(2*a+b, 2*a+(b+1), 8, 2))):
        x = (choice([0, 0.5, 1]))
        y = (choice([0, 0.5, 1]))
        pts.append([x, y])
    return Polyline(pts)
```

```python
widgets.interact(make_char,
                 a = i,
                 b = widgets.IntSlider(min=3, max=12, step=1, value=3))
```

```python
char_plines = [make_char(i) for i in (alpha_s)]
```
