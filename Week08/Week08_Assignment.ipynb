## Visualizing glyph complexity

As a continuation of the previous assignments, I took the lyrics from <i>The Rain In Spain</i> to visualize glyph complexity.

```python
def glyphcommands(f, ch):
    return Path([copy(cmd) for cmd in f.glyph(f.charmap[ord(ch)])])
def advancefor(f, ch):
    return f.advances[f.charmap[ord(ch)]]
```

```python
text1 = "The rain in Spain stays mainly in the plain"
text2 = "In Hertford, Hereford, and Hampshire, hurricanes hardly happen"
```

I started with making a slight variation on the code from https://github.com/aparrish/material-of-language/blob/master/manipulating-font-data.ipynb.

```python
# With text1: "The rain in Spain stays mainly in the plain" 
width = 360 # 5in
height = 220 # 4in
border = 9 # 1/8in
lineheight = 27 # size of each line
rectsize = 20 # size of each rectangle
page = document(width, height, 'pt').addpage()
yoff = border  
xoff = border
for line in text1_f:
    line = line.strip() # remove trailing whitespace
    for ch in line:
        path_len = len(glyphcommands(f, ch).commands)
        val = 255 - (path_len * 5)
        advance = advancefor(f, ch) / f.density * lineheight
        brush = shape().nostroke().fill(rgb(val, val, val))
        page.place(brush.rectangle(xoff, yoff, advance, rectsize))
        xoff += int(advance)
    xoff = border
    yoff += lineheight
show(page)
```

Then I decided to replace rectangular shapes with circles.

```python
# With text1: "The rain in Spain stays mainly in the plain" 

width = 360 # 5in
height = 300 # 4in
border = 24 # 1/8in
lineheight = 30 # size of each line
grid_size = 20
ellisize = width/grid_size
page = document(width, height, 'pt').addpage()
yoff = border  
for line in text1_f:
    for ch, i in zip(line, range(grid_size)):
        x = (i * ellisize) + ellisize * 0.5
        path_len = len(glyphcommands(f, ch).commands)
        val = 255 - (path_len * 5)
        advance = advancefor(f, ch) / f.density * lineheight
        brush = shape().nostroke().fill(rgb(val, val, val))
        page.place(brush.ellipse(x, yoff, ellisize*0.5, ellisize*0.5))
    yoff += lineheight
show(page)
```

```python
from itertools import chain
def combine_path(f, s):
    text_paths = []
    cx = 0
    for ch in s:
        glyph_path = glyphcommands(f, ch).translate(cx, 0)
        text_paths.append(glyph_path)
        cx += advancefor(f, ch)
    combined = Path(list(chain(*text_paths)))
    return combined
```

```python
def nudge(cmd, x, y):
    c = copy(cmd)
    if type(cmd) in (moveto, lineto, quadto, curveto):
        c.x += x
        c.y += y
    if type(cmd) in (quadto, curveto):
        c.x1 += x
        c.y1 += y
    if type(cmd) == curveto:
        c.x2 += x
        c.y2 += y
    return c
```

```python
s = "honey badger"
```

```python
def stretch_char(f, ch, height=-0.33):
    path_orig = glyphcommands(f, ch)
    path_cmds = []
    mid_y = f.ascender * height
    for cmd in path_orig:
        if hasattr(cmd, 'y') and cmd.y < mid_y:
            path_cmds.append(nudge(cmd, x=0, y=-200))
        else:
            path_cmds.append(cmd)
    path_mod = Path(path_cmds)
    return path_mod
```

```python
page = document(400, 90, 'pt').addpage()
pen = shape().nostroke().fill(rgba(240, 94, 35, 255))
factor = 72 / f.density
cx = 0
for ch in s:
    glyph_path = stretch_char(f, ch, uniform(-1, -0.25)).scale(factor).translate(cx, 90)
    page.place(pen.path(glyph_path))
    cx += advancefor(f, ch) * factor
show(page)
```
